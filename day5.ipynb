```python
df['maths']=[99,80,88,77]
print(df)
```

```python
df['grad']=df['score']>=80
print(df)
```

```python
df['score']=df['score']
print(df)
```

```python
def evenOdd(maths):
  if maths%2==0:
    return 'even'
  else:
    return 'odd'
```

```python
df['even or odd']=df['maths'].apply(evenOdd)
print(df)
```

```python
result={
    True:'pass',
    False:"fail"
}
df['p or f']=df['grad'].map(result)
print(df)
```

```python
df.dropna()
df.dropna(axis=1)
df.drop(thresh=3)
```

```python
df.drop('even or odd',axis=1,inplace=True)
print(df)
```

```python
df.rename(columns={'name':'student'},inplace=True)
print(df)
```

```python
df.replace("fail",np.nan,inplace=True)
print(df)
```

```python
df.sort_values(['score'], ascending=True, inplace=True)
print(df)
#single value sorting
```

```python
df.sort_values(['age','maths'],ascending=[False,True])
print(df)
#sorting multiple values it will not sort only single value sort will be appear
```
